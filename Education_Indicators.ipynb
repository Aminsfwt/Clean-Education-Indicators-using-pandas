#Import pandas library

```python
import pandas as pd
```

#First load first dataset Education Indecators


```python
#load data
df = pd.read_csv("Education_Indecators.csv")

print(df.head())
```

#Drop Country Code and Indicator Code columns

```python
df = df.drop(columns=['Country Code', 'Indicator Code'])
print(df.head())
```

#clean columns by lowercase names and replace spaces by undescore  

```python
#Clean up column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

#Take only years from 2000 to 2024 from the dataset and unpevot the dataset to make only column for the years and column for the indicators values

```python

# Identify year columns from 2000 to 2024
year_cols = [str(y) for y in range(2000, 2025) if str(y) in df.columns]

# Melt the dataset to convert years columns to one column
df_long = df.melt(
    id_vars=['country_name', 'indicator_name'],
    value_vars=year_cols,                       # columns to unpivot
    var_name='year',                            # new column name for years
    value_name='indicator_value'                # new column name for values
)

df_long.head()
```

#Convert year to integer and drop missing values

```python
df_long['year'] = df_long['year'].astype(int)
df_long.dropna(subset=['indicator_value'], inplace=True)
df_long.head()
```

```python
#remove impossible or extreme values
df_long.loc[df_long['indicator_value'] < 0, 'indicator_value'] = None
df_long.loc[df_long['indicator_value'] > 100, 'indicator_value'] = 100
```

#Save the cleaned and reshaped data


```python
df_long.to_csv("Education_Indicators.csv", index=False)
```
